```python
# Libraries

# General data handling
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import itertools

# Machine learning
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
# Reading in data
df = pd.read_csv("../data/processed_data.csv")
df.head()
```

```python
# Removing random first column
df = df.iloc[:,1:29]
df.head()
```

Now that we have our data, let's divide it into training and validation data. 

```python
X = df.iloc[:, 1:27]
y = df['label']
y = y.astype('category')

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.33)
```

## 1. Baseline Model

Let's first make a baseline logistic regression model to predict the different classes. From this model, we can examine any important features and residuals to note any interesting patterns

https://stats.stackexchange.com/questions/52104/multinomial-logistic-regression-vs-one-vs-rest-binary-logistic-regression

```python
log_clf = LogisticRegression().fit(X_train, y_train)
log_accuracy = log_clf.score(X_val, y_val)
log_confusion_matrix = confusion_matrix(y_val,log_clf.predict(X_val))
print("Log accuracy: ", log_accuracy)
```

Wow! This did terribly! This could be due to a few factors
1. The logistic model was not a good fit for this type of data: To correct this, I will try to work with other machine learning models and see if I can make any progress
2. The data is not great: This is most likely the problem with this project: I only have 20 songs per year, which may not be enough for my model to thoroughly train itself. To fix this, I will try to reduce the number of features to the ones that we found significant in our EDA and maybe even change our whole project scope so that we can get new data.

First, I would like to test if other models can possibly work better. I will first work around with the random forest model.

```python
rf_clf = RandomForestClassifier().fit(X_train, y_train)
rf_accuracy = rf_clf.score(X_val, y_val)
rf_confusion_matrix = confusion_matrix(y_val, rf_clf.predict(X_val))
print("Rf accuracy: ", rf_accuracy)
```

This pretty much shows that it is not likely to be a fault of logistic regression that the model did poorly: the random forest model did just as poorly. Let's see if we can see where the model is making its most misclassification by creating a confusion matrix

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(log_confusion_matrix, classes=['2010', '2011', '2012', '2013', '2014', '2015', '2016',
                                                    '2017', '2018'],
                      title='Logistic regression confusion matrix')
```

```python
plt.figure()
plot_confusion_matrix(rf_confusion_matrix, classes=['2010', '2011', '2012', '2013', '2014', '2015', '2016',
                                                    '2017', '2018'],
                      title='Random forest confusion matrix')
```

As we can see from the confusion matrices, it doesn't seem to be a particular class that is causing problems; rather, the model is just not the best when it comes to claissification of this sort.
