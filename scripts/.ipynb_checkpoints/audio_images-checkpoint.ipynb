# Audio to Images

I want to experiment with an image based approach to this problem instead of a tabular approach to data. We will use the librosa library to generate spectrograms of each audio clip and biuild a classifier basde on that. 

```python
import os
import librosa
import librosa.display
import numpy as np
import mathplotlib.pyplot as plt
```

Before we get ahead of ourselves, I want to test this feature on one audio clip

```python
clip, sample_rate = librosa.load('../music/english/2010s_audio/ChrisBrownDontWakeMeUpOfficialMusicVideoQOowQeKyNkQmp3_2.mp3')
clip, _ = librosa.effects.trim(clip)
librosa.display.waveplot(clip, sr = sample_rate)
```

```python
D = np.abs(librosa.stft(clip))
librosa.display.specshow(D, sr=sample_rate, x_axis='time', y_axis='linear')
```

```python
DB = librosa.amplitude_to_db(D, ref=np.max)
librosa.display.specshow(DB, sr=sample_rate, x_axis='time', y_axis='linear')
```

```python
S = librosa.feature.melspectrogram(clip, sr=sample_rate)
S_DB = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S_DB, sr=sample_rate)
```

Now, let's collect all the pictures and put them into the appropriate directories

```python
years = ['1970', '1980', '1990', '2000', '2010']

for year in years:
    subdir = f'../music/english/{year}s_audio'
    for filename in os.listdir(subdir):
        songname = f'../music/english/{year}s_audio/{filename}'
        save_path = f'../data/images/english/{year}s_audio/{filename}.png'
        clip, sample_rate = librosa.load(songname)
        
        D = np.abs(librosa.stft(clip))
        DB = librosa.amplitude_to_db(D, ref=np.max)
        S = librosa.feature.melspectrogram(clip, sr=sample_rate)
        S_DB = librosa.power_to_db(S, ref=np.max)
        
        pylab.axis('off') # no axis
        pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[]) # Remove the white edge
        
        librosa.display.specshow(S_DB, sr=sample_rate)
        
        pylab.savefig(save_path, bbox_inches=None, pad_inches=0)
        pylab.close()
```
