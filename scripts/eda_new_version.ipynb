### 1.2: Regression Testing
Why are we performing linear regression, and what does t-tests have to do with anything?

In this situation, we are uncertain if there is any relationship between years and spectral centroid, rolloff, zero crossing rate, chroma STFT, and spectral bandwidth

If we were certain, then we should be able to draw a linear regression line that has a non-zero slope. If there is no relationship, then the slope should be zero. 

This is what we are trying to do in this case. We will construct the a linear regression for each factor from 2010 to 2018. We will then examine the p-value of the slope, which will tell us the probability that the null hypothesis (that there is a slope of 0) is true given the data that we have. If the p-value is small, we can say with a reasonable amount of confidence that we are likely to have a non-zero slope, or in other words, an actual relationship. 

How much is a small p-value? To determine this, we will need to use Bonferroni correction as we are testing multiple hypotheses. Since we are using 3 tests, we will have a cut off value of 0.05/3 = 0.0167. 

Let's get started by creating a regression for rolloff.

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(data['rolloff'],data['year'])
print(f'The slope of the rolloff linear regression test is {slope}')
print(f'The p-value of the slope of the rolloff linear regression test is {p_value}')
print(f'The standard error of the slope of the rolloff linear regression test is {std_err}')
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(data['spectral_centroid'],data['year'])
print(f'The slope of the spectral centroid linear regression test is {slope}')
print(f'The p-value of the slope of the spectral centroid linear regression test is {p_value}')
print(f'The standard error of the slope of the spectral centroid linear regression test is {std_err}')
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(data['zero_crossing_rate'],data['year'])
print(f'The slope of the zero crossing rate linear regression test is {slope}')
print(f'The p-value of the slope of the zero crossing rate linear regression test is {p_value}')
print(f'The standard error of the slope of the zero crossing rate linear regression test is {std_err}')
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(data['chroma_stft'],data['year'])
print(f'The slope of the chroma STFT linear regression test is {slope}')
print(f'The p-value of the slope of the chroma STFT linear regression test is {p_value}')
print(f'The standard error of the slope of the chroma STFT linear regression test is {std_err}')
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(data['spectral_bandwidth'],data['year'])
print(f'The slope of the spectral bandwidth linear regression test is {slope}')
print(f'The p-value of the slope of the spectral bandwidth  linear regression test is {p_value}')
print(f'The standard error of the slope of the spectral bandwidth  linear regression test is {std_err}')
```

Looking at the results of the regression slope t-tests, we can be fairly confident that the slope between the zero crossing rate, chroma STFT and spectral bandwidth are not 0 as the p-value is well below our threshold of 0.0167.

### 1.3 MFCC 
One thing that we did not look at was the change in MFCC over the years. Before we do that, we should understand what each MFCC means.

The set of 20 MFCC coefficients are a way to describe the power spectrum of a signal across frequencies. This set of coefficients is often used for genre classification and other audio machine learning projects. Hence, it should be quite useful for this project. 

We will do the same process as last time for regression testing. Namely:
1. For each MFCC, construct a regression line from 2010 to 2020
2. Conduct a regression t-test
3. Loop through each MFCC coefficient

Since we are testing 20 hypotheses, we will perform Bonferroni correction. Thus, the p-value cutoff is 0.05/20 = 0.0025

```python
mfcc_list = []
for i in range (1, 21):
    mfcc_list.append(f'mfcc{i}')

for i in mfcc_list:
    slope, intercept, r_value, p_value, std_err = stats.linregress(data[i],data['year'])
    if p_value < 0.0025: # Only printing out significant MFCC for further analysis
        print(f'The slope of the {i} linear regression test is {slope}')
        print(f'The p-value of the slope of the {i} linear regression test is {p_value}')
        print(f'The standard error of the slope of the {i} linear regression test is {std_err}')
        print("************************************************************************************************")
```

Interesting! Almost all MFCC coefficients except 1, 2, and 10 passed the Bonferroni threshold

Let's create a few regression plots to visualize this.

```python
# MFCC3
plt.rcParams['figure.figsize'] = [20,10]
data['year'] = data['year'].astype('float64') # Just to get regression to work
sns.regplot(x="year", y="mfcc3", data = data);
```

## 2.2: Principal Component Analysis
Let's perform some principal component analysis to see if we can somehow reduce the number of features to a feww principal components that hopefully can explain a lot of the variance in the data.

```python
x = data.iloc[:,1:27].values
y = data['year']

x = StandardScaler().fit_transform(x)

pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])
principalDf = pd.concat([principalDf, y], axis = 1)

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 


ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [70, 80, 90, 0, 10, 2015, 2016, 2017, 2018]
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k', 'burlywood', '#68A3B3']
for target, color in zip(targets,colors):
    indicesToKeep = principalDf['label'] == target
    ax.scatter(principalDf.loc[indicesToKeep, 'PC1']
               , principalDf.loc[indicesToKeep, 'PC2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
ax.plot() 
```

```python
print(pca.explained_variance_ratio_)
```

Looking at the graph and the percentage of variation that is explained by the principal components, we can conclude that reducing the variables down to principal components has noticeable benefit in explaining the data in a better way. 

However, this provides us a conundrum: our heatmap and analysis into the Pearson's correlation coefficient showed that some variables are highly correlated, while our PCA plot shows that is not the case. This is most probably due to the effect of years: some years may have high correlation between variables, while others are not. Overall, this would lead to a high correlation, masking the years where the correlation is not as high.
